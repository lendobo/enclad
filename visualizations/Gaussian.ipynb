```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for Gaussian bell curve (mean and standard deviation)
mu = 0
sigma = 2

# Generate data points
x = np.linspace(-7, 7, 400)
y = (1/(sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu)/sigma)**2)

# Plotting the Gaussian bell curve
plt.figure(figsize=(12, 6), dpi = 300)
plt.plot(x, y, 'grey', linewidth=3)
# remove x ticks 
plt.grid(False)
# remove top and right spines
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
# remove x and y ticks
plt.xticks([])
plt.yticks([])
plt.ylabel('Probability Density', fontsize=20)
plt.ylim(0, 0.21)

plt.annotate(r'$\mu^*_k$', xy=(0.8, 0.3), xycoords='axes fraction', fontsize=25, color='#a10000')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters for binomial distribution
n = 50  # number of trials
p = 0.5  # probability of success

# Generate data
x = np.arange(0, n+1)
pmf = np.bincount(np.random.binomial(n, p, 10000)) / 10000  # Approximating 

# Correctly compute the PMF using scipy.stats

pmf = binom.pmf(x, n, p)

# Plot
plt.figure(figsize=(12, 6), dpi = 300)
plt.bar(x, pmf, width=0.7, align='center', color='grey', alpha=0.7)
# plt.xlabel('Number of Successes')
plt.ylabel('Probability Mass', fontsize=20)
# plt.title('Binomial Distribution: n=50, p=0.5')
plt.xlim(10, 40)
plt.grid(False)
# remove top and right spines
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
# remove x and y ticks
plt.xticks([])
plt.yticks([])

# generate a label at y = 0.3 and x = 0.8
plt.annotate(r'$P^p_k \cdot Q$', xy=(0.8, 0.3), xycoords='axes fraction', fontsize=25, color='#a10000')
plt.annotate(r'$z_k^{\lambda^j}$', xy=(0.95, 0.3), xycoords='axes fraction', fontsize=25, color='black')

# plt.tight_layout()

plt.show()
```
